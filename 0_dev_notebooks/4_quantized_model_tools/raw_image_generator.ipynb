```python
OUTPUT_DIR = "./raw_images/"
```

```python
import cv2
import os
import natsort
```

```python
test_images = list()

for root, dir, files in os.walk("../../dataset/resize_128/test/"):
    for file in files:
        test_images.append(os.path.join(root, file))

test_images = natsort.natsorted(test_images)
print(len(test_images))
```

```python
def get_name(path: str) -> str:
    name = path.split(os.sep)[-1]
    name = name.replace(' ', '_')
    name = name.replace("(", "")
    name = name.replace(")", "")
    name = name.split('.')[0]
    return name
```

```python
os.makedirs(OUTPUT_DIR, exist_ok=True)

for img_path in test_images:
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    name = get_name(img_path)
    print(name)
    with open(os.path.join(OUTPUT_DIR, name + ".raw"), 'bw') as file:
        file.write(img)
```
